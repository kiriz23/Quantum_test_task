# Dataset creation

For generation of text I used ChatGPT, then asked it list of all mentioned mountains. 

```python
# Sample sentences from ChatGPT
sentences = [
    "Mount Everest is the highest peak in the world.",
    "K2 is known as the Savage Mountain.",
    "The Alps are famous for their breathtaking landscapes and winter sports.",
    "Mount Godwin-Austen , often called K2 , is a challenging peak to climb.",
    "Climbers from around the world gather to ascend Mount Everest each year.",
    "Mount Kilimanjaro is located in Tanzania and is Africa's highest peak.",
    "The Himalayas are a majestic mountain range in Asia.",
    "Denali, formerly known as Mount McKinley, is the highest peak in North America.",
    "The Andes run through several South American countries, including Chile and Peru.",
    "The Rocky Mountains are a famous range in North America, stretching from Canada to New Mexico.",
    "Mount Fuji is a sacred mountain in Japan and a popular tourist attraction.",
    "The Pyrenees form a natural border between Spain and France.",
    "Climbers dream of reaching the summit of Mount Elbrus, the highest peak in Europe.",
    "The Karakoram range, home to K2 , is renowned for its rugged terrain and high peaks.",
    "Aconcagua is the tallest mountain in South America and is part of the Andes.",
    "Mount Everest is the highest peak in the world and attracts climbers from all over the globe.",
    "The Alps stretch across several countries in Europe, offering stunning views and great skiing opportunities.",
    "Everest Base Camp is a popular trekking destination for adventurers.",
    "Many mountaineers consider K2 the ultimate test of their skills and endurance.",
    "The Alps are known for their picturesque villages and vibrant wildlife."
]

# List of known mountain names
mountain_names = ["Mount Everest", "K2", "The Alps", "Mount Godwin-Austen", "Savage Mountain", "Kilimanjaro", "Himalayas", "Denali", "Mount McKinley", "Andes", 
                 "Rocky Mountains", "Mount Fuji", "Pyrenees", "Mount Elbrus", "Karakoram Range", "Aconcagua", "Karakoram"
                 ]
```

### Labeling mountains
Function below tokenizes text and annotates all mountain names in it, using BIO format.<br>
Labels used: <br>
***B-MOUNTAIN***: Beginning of a mountain name.<br>
***I-MOUNTAIN***: Inside a multi-word mountain name.<br>
***O***: Outside any entity.<br>

```python
# Function to tokenize and annotate text
def create_bio_tags(sentences, mountain_names):
    bio_dataset = []
    for sentence in sentences:
        words = sentence.split()  # Simple tokenization
        labels = ["O"] * len(words)  # Default label is "O"

        for mountain in mountain_names:
            mountain_tokens = mountain.split()
            for i in range(len(words) - len(mountain_tokens) + 1):
                # Check if tokens match a mountain name
                if words[i:i+len(mountain_tokens)] == mountain_tokens:
                    labels[i] = "B-MOUNTAIN"  # First token of the mountain name
                    for j in range(1, len(mountain_tokens)):
                        labels[i+j] = "I-MOUNTAIN"  # Remaining tokens

        bio_dataset.append(list(zip(words, labels)))
    return bio_dataset

# Create the BIO-tagged dataset
bio_dataset = create_bio_tags(sentences, mountain_names)

# Print the dataset in BIO format
for sentence in bio_dataset:
    for word, label in sentence:
        print(f"{word} {label}")
    print("")  # Blank line between sentences
```

### Saving the Dataset

```python
with open("D:\\InterviewProject\\Task1\\dataset\\dataset_bio.txt", "w") as f:
    for sentence in bio_dataset:
        for word, label in sentence:
            f.write(f"{word} {label}\n")
        f.write("\n")
```
